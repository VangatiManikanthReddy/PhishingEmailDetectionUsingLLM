## Zeroshot

```python
# Zero shot prompt for the header
def create_zeroshot_prompt(row, email_type="header"):

    if email_type == "header":
        return f"""
Classify this email header as 'Phishing' or 'Legitimate'. Then briefly explain why.

Sender: {row['sender']}
Receiver: {row['receiver']}
Subject: {row['subject']}
Date: {row['date']}
i need both the prediction and the reason for every thing
Format: Phishing or Legitimate|Reason
"""

    elif email_type == "body":
        return f"""
Classify this email body as 'Phishing' or 'Legitimate'. Then briefly explain why.

Body: {row['body']}

Format: Phishing or Legitimate|Reason
"""

    elif email_type == "combined":
        return f"""
Classify this email (header and body) as 'Phishing' or 'Legitimate'. Then briefly explain why.

Header:
Sender: {row['sender']}
Receiver: {row['receiver']}
Subject: {row['subject']}
Date: {row['date']}

Body:
{row['body']}

Format: Phishing or Legitimate|Reason
"""

    else:
        raise ValueError(" Invalid email_type specified.")
```

```python
import requests

def call_llama(prompt):
    url = "https://api.llama-api.com/chat/completions"
    headers = {
        "Content-Type": "application/json",
        "Authorization": f"Bearer {llama_api_key}"
    }
    payload = {
        "model": "llama3.1-70b",
        "messages": [{"role": "user", "content": prompt}]
    }

    try:
        response = requests.post(url, headers=headers, json=payload)
        response.raise_for_status()  # Raise HTTPError for bad responses (4xx or 5xx)
        result = response.json()["choices"][0]["message"]["content"].strip()
        print(result)

        # Split the result by '|' but ensure we always have 2 elements
        parts = result.split("|", 1)  # Split at the first '|' only
        if len(parts) < 2:
            parts.append("No reason provided.")  # Add a default reason if missing

        return parts  # Return the list

    except (requests.exceptions.RequestException, KeyError, IndexError) as e:
        print(f"❌ Error calling Llama API: {e}")  # Print error for debugging
        return ["❌ API Error", ""]  # Return error values for both columns
```

```python
# Processes a CSV file from sample_75 and saves predictions to a new Excel file in output_data
def process_sample_75_csv_to_output(file_name, email_type):
    # Construct full path to input CSV file
    input_path = os.path.join(path,'001 Submission','All APIs', 'llama_llm', 'prompt_data','sample_75_without_labels',file_name)
        # input_path = f'/content/drive/MyDrive/outPut_Folder/gpt_llm/prompt_data/sample_75_without_labels/{file_name}'
    # Define output folder location
    output_folder = os.path.join(path,'001 Submission','All APIs', 'llama_llm', 'output_daata','zeroshot')
    #  '/content/drive/MyDrive/outPut_Folder/gpt_llm/output_daata/zeroshot'
    os.makedirs(output_folder, exist_ok=True)

    # Construct output filename by replacing '.csv' with '_predicted.xlsx'
    output_file = file_name.replace('.csv', '_predicted.xlsx')
    output_path = os.path.join(output_folder, output_file)

    #  Load the CSV input into a pandas DataFrame
    df = pd.read_csv(input_path)


    #  Apply GPT to each row using zero-shot prompt
    df[['Prediction', 'Reason']] = df.apply(
        lambda row:pd.Series(call_llama(create_zeroshot_prompt(row, email_type))),
        axis=1
    )


    # Convert predictions to numeric labels
    df["prediction_label"] = df["Prediction"].map(lambda x: 1 if x.strip().lower() == "phishing" else 0)

    # Reorder prediction_label column right after Prediction
    cols = df.columns.tolist()
    if "Prediction" in cols and "Reason" in cols and "prediction_label" in cols:
        pred_idx = cols.index("Prediction")
        cols.insert(pred_idx + 1, cols.pop(cols.index("prediction_label")))
        df = df[cols]

    #  Save the final DataFrame as Excel file
    df.to_excel(output_path, index=False)
    print(f"✅ Processed: {file_name} → Saved to {output_path}")

```

## Fewshot

```python
# Pre-defined examples (few-shot) to "teach" the model before each real classification
FEW_SHOT_EXAMPLES = {
    "header": [
        {
            "sender": "admin@securebank.com",
            "receiver": "user123@gmail.com",
            "subject": "Important Account Notification",
            "date": "2024-11-05",
            "label": "Phishing",
            "reason": "The sender's email is suspicious and urges action."
        },
        {
            "sender": "newsletter@trustedsource.com",
            "receiver": "user123@gmail.com",
            "subject": "Monthly Newsletter - November",
            "date": "2024-11-01",
            "label": "Legitimate",
            "reason": "It’s a regular newsletter from a known sender."
        }
    ],
    "body": [
        {
            "body": "Your account has been locked. Click here to verify immediately: http://scam-site.com",
            "label": "Phishing",
            "reason": "Urgent tone and suspicious link included."
        },
        {
            "body": "Here is your delivery confirmation for order #12345. Thank you for shopping with us!",
            "label": "Legitimate",
            "reason": "Normal e-commerce confirmation message."
        }
    ],
    "combined": [
        {
            "sender": "alerts@paypal-security.com",
            "receiver": "user@gmail.com",
            "subject": "Verify your PayPal account now!",
            "date": "2024-11-08",
            "body": "We noticed suspicious activity. Click below to confirm.",
            "label": "Phishing",
            "reason": "Suspicious sender and urgent call-to-action."
        },
        {
            "sender": "support@github.com",
            "receiver": "user@gmail.com",
            "subject": "New login detected",
            "date": "2024-11-01",
            "body": "We noticed a new login from a different location. If this was you, no further action is required.",
            "label": "Legitimate",
            "reason": "Clear explanation, no suspicious links, from a known service."
        }
    ]
}
```

```python
def create_few_shot_prompt(row, email_type="header"):
    examples = FEW_SHOT_EXAMPLES[email_type]
    prompt = "You are an email security classifier. Learn from the examples below.\n\n"

    # Add compact few-shot examples
    for ex in examples:
        if email_type == "header":
            prompt += (
                f"Sender: {ex['sender']}, Subject: {ex['subject']}, Date: {ex['date']}\n"
                f"Answer: {ex['label']} | {ex['reason']}\n\n"
            )
        elif email_type == "body":
            prompt += (
                f"Body: {ex['body']}\n"
                f"Answer: {ex['label']} | {ex['reason']}\n\n"
            )
        elif email_type == "combined":
            prompt += (
                f"Sender: {ex['sender']}, Subject: {ex['subject']}, Date: {ex['date']}, Body: {ex['body']}\n"
                f"Answer: {ex['label']} | {ex['reason']}\n\n"
            )

    # Now add the row to classify
    if email_type == "header":
        prompt += (
            f"Sender: {row['sender']}, Subject: {row['subject']}, Date: {row['date']}\n"
            f"Answer:"
        )
    elif email_type == "body":
        prompt += (
            f"Body: {row['body']}\n"
            f"Answer:"
        )
    elif email_type == "combined":
        prompt += (
            f"Sender: {row['sender']}, Subject: {row['subject']}, Date: {row['date']}, Body: {row['body']}\n"
            f"Answer:"
        )

    return prompt
```

```python
def process_sample_75_few_shot(file_name, email_type):
    input_path = os.path.join(path,'001 Submission','All APIs', 'llama_llm', 'prompt_data','sample_75_without_labels',file_name)
    # f'/content/drive/MyDrive/outPut_Folder/gpt_llm/prompt_data/sample_75_without_labels/{file_name}'
    output_folder = os.path.join(path,'001 Submission','All APIs', 'llama_llm', 'output_daata','fewshot')
    #  '/content/drive/MyDrive/outPut_Folder/gpt_llm/output_daata/fewshot'
    os.makedirs(output_folder, exist_ok=True)

    output_file = file_name.replace('.csv', '_fewshot_predicted.xlsx')
    output_path = os.path.join(output_folder, output_file)

    # Load input data
    df = pd.read_csv(input_path)

    # Apply GPT to each row using few-shot prompt
    df[['Prediction', 'Reason']] = df.apply(
        lambda row:pd.Series(call_llama(create_few_shot_prompt(row, email_type))),
        axis=1
    )

    # Convert textual label to numeric
    df["prediction_label"] = df["Prediction"].map(lambda x: 1 if x.strip().lower() == "phishing" else 0)

    # Reorder prediction_label
    cols = df.columns.tolist()
    if "Prediction" in cols and "Reason" in cols and "prediction_label" in cols:
        pred_idx = cols.index("Prediction")
        cols.insert(pred_idx + 1, cols.pop(cols.index("prediction_label")))
        df = df[cols]

    # Save to Excel
    df.to_excel(output_path, index=False)
    print(f"✅ Few-shot processed and saved → {output_path}")
```

## Chain of thought

```python
def create_chain_of_thought_prompt(row, email_type="header"):
    if email_type == "header":
        return f"""
You are a cybersecurity expert specializing in phishing detection.
Carefully review the email header provided and predict whether it is 'Phishing' or 'Legitimate'.

Instructions:
- Analyze sender, receiver, subject, and date briefly.
- Predict if the email is Phishing or Legitimate.
- Directly output in the format given below without any detailed explanation.

Header:
Sender: {row['sender']}
Receiver: {row['receiver']}
Subject: {row['subject']}
Date: {row['date']}
Compare the instructions or requests made in the email with the sender information, subject line, and other header details.
If **any one part of the instruction seems suspicious, misleading, or abnormal**, classify the email as "Phishing".
You must return only one label: either Phishing or Legitimate. Do not return both.
Output Format:
Phishing or Legitimate | Reason: [Short reason here]
(❗ No need for step-by-step analysis. Only final short prediction and reason.)
"""

    elif email_type == "body":
        return f"""
You are a cybersecurity expert specializing in phishing detection.
Carefully review the email body provided and predict whether it is 'Phishing' or 'Legitimate'.

Instructions:
- Analyze the email body briefly.
- Predict if the email is Phishing or Legitimate.
- Directly output in the format given below without any detailed explanation.

Body:
{row['body']}
Compare the instructions or requests made in the email with the sender information, subject line, and other header details.
If **any one part of the instruction seems suspicious, misleading, or abnormal**, classify the email as "Phishing".
You must return only one label: either Phishing or Legitimate. Do not return both.
Output Format:
Phishing or Legitimate | Reason: [Short reason here]
(❗ No need for step-by-step analysis. Only final short prediction and reason.)
"""

    elif email_type == "combined":
        return f"""
You are a cybersecurity expert specializing in phishing detection.
Carefully review the full email (header and body) and predict whether it is 'Phishing' or 'Legitimate'.

Instructions:
- Analyze sender, receiver, subject, date, and body briefly.
- Predict if the email is Phishing or Legitimate.
- Directly output in the format given below without any detailed explanation.

Header:
Sender: {row['sender']}
Receiver: {row['receiver']}
Subject: {row['subject']}
Date: {row['date']}

Body:
{row['body']}

Compare the instructions or requests made in the email with the sender information, subject line, and other header details.
If **any one part of the instruction seems suspicious, misleading, or abnormal**, classify the email as "Phishing".
You must return only one label: either Phishing or Legitimate. Do not return both.
Output Format:
Phishing or Legitimate | Reason: [Short reason here]
(❗ No need for step-by-step analysis. Only final short prediction and reason.)
"""

    else:
        raise ValueError("❌ Invalid email_type passed to prompt generator.")
```

```python
# ✅ Loads a sample_75 CSV file, applies CoT prompting, and saves results as Excel
def process_chain_of_thought(file_name, email_type="header"):
    #  Define paths
    input_path = os.path.join(path,'001 Submission','All APIs', 'llama_llm', 'prompt_data','sample_75_without_labels',file_name)
    #  f'/content/drive/MyDrive/outPut_Folder/gpt_llm/prompt_data/sample_75_without_labels/{file_name}'
    output_folder = os.path.join(path,'001 Submission','All APIs', 'llama_llm', 'output_daata','chainofthought')
    #  '/content/drive/MyDrive/outPut_Folder/gpt_llm/output_daata/chainofthought'
    os.makedirs(output_folder, exist_ok=True)

    output_file = file_name.replace('.csv', '_cot_predicted.xlsx')
    output_path = os.path.join(output_folder, output_file)

    # Load input CSV into DataFrame
    df = pd.read_csv(input_path)

    #  Generate CoT prompt per email and apply GPT-4o
    df[['Prediction', 'Reason']] = df.apply(
        lambda row:pd.Series(call_llama(create_chain_of_thought_prompt(row, email_type))),
        axis=1
    )


    # Map textual prediction to numeric label (1 = phishing, 0 = legitimate)
    df["prediction_label"] = df["Prediction"].apply(
        lambda x: 1 if isinstance(x, str) and "phishing" in x.lower() else 0
    )

    #  Reorder columns (optional)
    cols = df.columns.tolist()
    if "Prediction" in cols and "Reason" in cols and "prediction_label" in cols:
        pred_idx = cols.index("Prediction")
        cols.insert(pred_idx + 1, cols.pop(cols.index("prediction_label")))
        df = df[cols]

    # 🔹 Save final DataFrame as Excel
    df.to_excel(output_path, index=False)

    # ✅ Confirm completion
    print(f"✅ CoT analysis complete for '{email_type}' emails. File saved at: {output_path}")
```
